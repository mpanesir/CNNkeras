```python
# Importing libraries
```

```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation, Flatten
from keras.layers.convolutional import Convolution2D, MaxPooling2D
from sklearn.model_selection import train_test_split
```

```python
# Importing the MNIST data
```

```python
data = pd.read_csv("mnist.csv")
```

```python
# Checking the head of the dataset
```

```python
data.head()
```

```python
# Reshaping our data to a 28*28 array for the CNN to work efficeintly
```

```python
data.iloc[3,1:].values.reshape(28,28).astype('uint8')
```

```python
# Performing data preprocessing 
```

```python
df_x = data.iloc[:,1:].values.reshape(len(data),28,28,1)
```

```python
# Storing labels in y
y = data.iloc[:,0].values
```

```python
# Converting the labels into categorical features using to_categorical function of Keras
```

```python
df_y = keras.utils.to_categorical(y, num_classes=10)
```

```python
# Converting df_x and df_y into numpy array for the train_test_split
```

```python
df_x = np.array(df_x)
df_y = np.array(df_y)
```

```python
# Train Test Split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.33)
```

```python
#Genrating a Keras CNN Model :
```

```python
model = Sequential()
model.add(Convolution2D(32,3,data_format="channels_last", activation="relu", input_shape = (28,28,1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(100))
model.add(Dropout(0.5))
model.add(Dense(10))
model.add(Activation("softmax"))
model.compile(loss = "categorical_crossentropy", optimizer = "adadelta", metrics = ["accuracy"])
```

```python
# Model Summary
```

```python
model.summary()
```

```python
# Fitting the model for test
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test), epochs=20)
```

```python
model.evaluate(x_test, y_test)
```
